### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
players_count=pd.DataFrame({"Total Number of Players":[len(purchase_data)]})
players_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=len(purchase_data["Item ID"].value_counts())
average_price=round(purchase_data["Price"].mean(),2)
number_purchase=len(purchase_data)
tot_rev=purchase_data["Price"].sum()
P_Analysis=pd.DataFrame({
    "Number of Unique Items":[unique_items],
    "Average Price":[f"${average_price}"],
    "Number of Purchases":[number_purchase],
    "Total Revenue":[f"${tot_rev}"]
})
P_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Gender_demo=pd.DataFrame(purchase_data["Gender"].value_counts())
Gender_demo["Percentage of Players"]=round((purchase_data["Gender"].value_counts()/purchase_data["Gender"].count())*100,2).astype(str)+"%"
Gender_demo=Gender_demo.rename(columns={"Gender": "Total Count"})
Gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
count=pd.DataFrame(purchase_data.groupby("Gender")["Purchase ID"].count())
price_avg=purchase_data.groupby("Gender")["Price"].mean()
purchase_value=purchase_data.groupby("Gender")["Price"].sum()
#purchase_data.groupby("Gender")["Price"].sum()/purchase_data.groupby("Gender")["Purchase ID"].count()
purchase_value
```

```python
x=purchase_data.groupby(["SN","Gender"])["Price"].sum()
avg_ppp=x.groupby("Gender").mean()
```

```python

purchase_a=count
purchase_a["Average Purchase Price"]=price_avg.astype(float).map("${:.2f}".format)
purchase_a["Total Purchase Value"]=purchase_value.astype(float).map("${:.2f}".format)
purchase_a["Avg Total Purchase p Person"]=avg_ppp.astype(float).map("${:.2f}".format)
purchase_a=purchase_a.rename(columns={"Purchase ID":"Purchase Count"})
purchase_a
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,80]
bin_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Groups"]=pd.cut(purchase_data["Age"],bins,labels=bin_labels)
age_group=pd.DataFrame(purchase_data.groupby("Age Groups")["Age Groups"].count())
age_group["Percentage of Players"]=round((age_group["Age Groups"]/len(purchase_data))*100,2).astype(str)+"%"
age_group=age_group.rename(columns={"Age Groups":"Total Count"})
age_group
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_a=pd.DataFrame(purchase_data.groupby("Age Groups")["Age"].count())
age_a
```

```python
age_a["Average Purchase Price"]=purchase_data.groupby("Age Groups")["Price"].mean().astype(float).map("${:.2f}".format)
age_a["Total Purchase Value"]=purchase_data.groupby("Age Groups")["Price"].sum().astype(float).map("${:.2f}".format)
age_a
```

```python
z=purchase_data.groupby(["Age Groups","SN"])["Price"].sum()
age_a["Avg Total Purcahse per Person"]=z.groupby("Age Groups").mean().astype(float).map("${:.2f}".format)
age_a=age_a.rename(columns={"Age":"Purchase Count"})
age_a
```
